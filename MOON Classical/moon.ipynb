```python
import pandas as pd
from sklearn.datasets import make_moons
```

```python
# Generate a synthetic dataset using make_moons
X, y = make_moons(n_samples=500, noise=0.1, random_state=42)

# Convert the dataset into a pandas DataFrame
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Target'] = y

# Display the first few rows of the DataFrame
print(df.head())
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
# Visualize the dataset
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Feature1', y='Feature2', hue='Target', palette='Set1', s=100)
plt.title('MOON Dataset Visualization', fontsize=16)
plt.xlabel('Feature 1', fontsize=14)
plt.ylabel('Feature 2', fontsize=14)
plt.legend(title='Target', fontsize=12)
plt.show()

```
